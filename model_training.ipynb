```python
ratings = pd.read_csv("./dataset/BX-Book-Ratings.csv", sep=";", encoding="latin-1")

ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={"User-ID": "user_id", "Book-Rating": "rating"}, inplace=True)
```

```python
ratings["user_id"].unique().shape[0]  # No of users
```

```python
x = ratings["user_id"].value_counts() > 200
```

```python
y = x[x].index  # Users ID that have written more than 200 reviews
```

```python
ratings = ratings[ratings["user_id"].isin(y)]
```

```python
ratings
```

```python
ratings.shape
```

```python
ratings.duplicated().value_counts()
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
ratings_with_books.shape
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby("title")["rating"].count().reset_index()
```

```python
number_rating.rename(columns={"rating": "no_of_ratings"}, inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on="title")
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating["no_of_ratings"] >= 50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(["user_id", "title"], inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(
    columns=["user_id"], index="title", values="rating"
)
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(metric="cosine", algorithm="brute")
```

```python
model.fit(book_sparse)
```

```python
model.get_params()
```

```python
distances, suggestions = model.kneighbors(
    book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6
)
```

```python
suggestions
```

```python
for i in suggestions:
    print(book_pivot.index[i])
```
